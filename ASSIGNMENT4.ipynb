```python
gd= []
def gradient_descent(start, learning_rate, number_of_itrs):
    x= start
    gd.append(start)
    for i in range(number_of_itrs):
        grad = f_dervative(x)
        x = x - learning_rate * grad
        gd.append(x)
        print(f'for iteration {i + 1}: x= {x}, f(x)= {f(x)}')\
        
    return x
```
