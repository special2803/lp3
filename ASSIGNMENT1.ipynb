```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('uber.csv')
df.head(10)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.drop(['Unnamed: 0','key', 'pickup_datetime'], axis= 1, inplace= True)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace= True)
df.isnull().sum()
```

```python
fig, axes = plt.subplots(len(df.columns), 1, figsize= (10, 6* len(df.columns)))
fig.tight_layout(pad= 3.0)

for i, col in enumerate(df.columns):
    sns.boxplot(data= df[col], ax= axes[i])
    axes[i].set_title(f'boxplot for {col}')
    axes[i].set_xlabel('')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data= df)
```

```python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lb = Q1 - 1.5 * IQR
    ub = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lb) & (df[col] <= ub)]

    

    
plt.figure(figsize=(10,6))
sns.boxplot(data= df)
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df
```

```python
sns.heatmap(data = df.corr(), annot= True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
X= df.drop(columns=['fare_amount'], axis= 1)
Y= df['fare_amount']

X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.25)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
Y_pred= linear_model.predict(X_test)
```

```python
r2score = r2_score(Y_test, Y_pred)
print(f'The R2 score is: {r2score}')
```

```python
rmse = np.sqrt(mean_squared_error(Y_pred, Y_test))
print(f'the RMSE is: {rmse}')
```

```python
from sklearn.ensemble import RandomForestRegressor
randomforest_model = RandomForestRegressor()
randomforest_model.fit(X_train, Y_train)
```

```python
rfr_Y_pred = randomforest_model.predict(X_test)
```

```python
r2score_rf = r2_score(Y_test, rfr_Y_pred)
print(f'The R2 score is: {r2score_rf}')
rmse_rf = np.sqrt(mean_squared_error(Y_test, rfr_Y_pred))
print(f'the RMSE is: {rmse_rf}')
```
